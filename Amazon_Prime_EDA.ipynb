## 3. ***Data Wrangling***

```python
# Making dataset analysis ready.

data.isnull().sum() # checking for null/missing values
```

```python
# Performing Constant/Placeholder Imputation on description column
data['description']=data['description'].fillna("No Description available")
```

```python
# Replacing all null values of Age certification column with mode( most frequently appearing value)
mode_value=data['age_certification'].mode()[0]
data['age_certification']=data['age_certification'].fillna(mode_value)
```

```python
# Replacing all null values in season column with 0
data['seasons']=data['seasons'].fillna(0)
```

```python
# Dropping null values from imdb_id as imdb_id is unique for movies and TV shows
data=data.dropna(subset=['imdb_id'])
```

```python
# Replacing null values in imdb_score column with mean of imdb_score
mean_value=data['imdb_score'].mean()
data['imdb_score']=data['imdb_score'].fillna(mean_value)
```

```python
# Replacing null values in imdb_votes column with 0
data['imdb_votes']=data['imdb_votes'].fillna(0)
```

```python
# Replacing null values in tmdb_popularity column with mean of tmdb_popularity
data['tmdb_popularity']=data['tmdb_popularity'].fillna(data['tmdb_popularity'].mean())
```

```python
# Replacing null values in tmdb_score column with mean of tmdb_score
mean_val=data['tmdb_score'].mean()
data['tmdb_score']=data['tmdb_score'].fillna(mean_val)
```

```python
# Replacing null values in character column with 'Unknown' as we don't know who played the character
data['character']=data['character'].fillna('Unknown')
```

 ## **1. Univariate Analysis**

#### **Chart - 1 : Count plot**

## 1. Content Type Distribution ( Movies vs TV shows)



```python
plt.figure(figsize=(8,6))
ax = sns.countplot(data=data, x='type')
for p in ax.patches:
    ax.annotate(
        p.get_height(),
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='bottom'
    )
plt.title("Content Type Distribution")
plt.xlabel("Content Type")
plt.ylabel("Count")
plt.show()
```

**Why this chart:** To understand content composition

**Insight:** Amazon Prime have more movies than TV shows ( Movie is dominating)

**Business Impact:** Helps decide where to invest more (movies or series)
